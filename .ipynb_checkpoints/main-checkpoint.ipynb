```python
# import required packages
# if pyspark is not installed - pip install pyspark
import os
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import folium
import folium.plugins as plugins

pd.set_option('display.max_columns',None)
```

```python
# Check if the file exists or not -> file name Global_YouTube_Statistics.csv
if os.path.exists("Global_YouTube_Statistics.csv"):
    yt = pd.read_csv("Global_YouTube_Statistics.csv",encoding="latin-1")
    print("File read successfully!!")
else:
    print("File doesn't exist")
```

```python
#separating numerical and categorical features
categorical_features=[feature for feature in yt.columns if yt[feature].dtype=='object']
numerical_features=[feature for feature in yt.columns if feature not in categorical_features]
yt_num=yt[numerical_features]
yt_cat=yt[categorical_features]
```

```python
#checking the top ten youtuber based on views
sns.set_style('darkgrid')
plt.figure(figsize=(10,6))
ax=sns.barplot(y='video views',x='Youtuber',data=yt,order=yt.sort_values('video views',ascending=False).Youtuber.iloc[:10],palette='magma')
plt.xticks(rotation=90)
ax.set_yticklabels([f'{int(label/1e9)}B' for label in ax.get_yticks()])
plt.title('Top 10 Youtubers in terms of Views(in billions)')
```

```python
#correlation heatmap
plt.figure(figsize=(25,25))
corr=yt_num.corr()
msk = np.triu(np.ones_like(corr))
sns.heatmap(corr,cmap=plt.cm.RdBu_r,annot=True,annot_kws={'size':10},mask=msk)
```

```python
# checking the number of youtube channel based on country
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(16,6))
sns.countplot(x='category',data=yt,order=yt['category'].value_counts(ascending=False).index,palette='magma',ax=ax[0])
sns.countplot(x='Country',data=yt,order=yt['Country'].value_counts(ascending=False).index,palette='magma',ax=ax[1])
ax[0].set_xticklabels(list(yt['category'].value_counts(ascending=False).index),rotation=90)
ax[1].set_xticklabels(list(yt['Country'].value_counts(ascending=False).index),rotation=90)
ax[0].set_title('No. of Youtube channels from each category')
ax[1].set_title('No. of youtube channels from each country')
```

```python
#the relations will be returned to the square graph
f, ax = plt.subplots(figsize=(15,15)) #subgraph is created by using figsize method to decide the weight and height.
sns.heatmap(yt.corr(), annot = True, linewidths=0.5,fmt ='0.1f', ax = ax) #heatmap graph is designed by using the specific parameters.
```

```python
top_10 = yt.head(10).sort_index(ascending=False)

# Creating a horizontal bar plot for the top 10 channels based on subscriber count.
top_10.plot(
    kind="barh", 
    x="Youtuber", 
    y=["subscribers"], 
    color="#c4302b", 
    legend=False, 
    figsize=(10, 5), 
    edgecolor="black", 
    linewidth=2
)

# Labeling the x-axis.
plt.xlabel("number of subscribers")

# Displaying the plot.
plt.show()
```

```python
# Dictionary mapping each channel type to a distinct color for visual consistency in plots.
color_dict = {
    'Entertainment': '#E63946',
    'Music': '#F1FAEE',
    'People': '#A8DADC',
    'Games': '#457B9D',
    'Comedy': '#F4A261',
    'Education': '#2A9D8F',
    'Film': '#264653',
    'Howto': '#E9C46A',
    'News': '#F77F00',
    'Tech': '#2B2D42',
    'Sports': '#D90429',
    'Autos': '#8D99AE',
    'Animals': '#81B29A',
    'Nonprofit': '#3D405B'
}

# Get the counts of each unique channel type from the dataset
type_counts = yt["channel_type"].value_counts()

# Plot a bar chart representing the counts of each channel type
type_counts.plot(kind="bar",figsize=(13, 6), 
                color=[color_dict[x] for x in type_counts.index],
                )

# Providing axis labels
plt.ylabel("Number of channels")
plt.xlabel("Channel type")

# Providing chart title
plt.title("Count of channels by their type")
```

```python
# Grouping the YouTube dataset by 'channel_type' and calculating the average number of subscribers for each type
subscribers_by_channel = yt.groupby("channel_type")["subscribers"].mean().sort_values(ascending=False)

# Plotting the average number of subscribers for each channel type in a bar chart
subscribers_by_channel.plot(kind="bar",figsize=(13, 6), 
                color=[color_dict[x] for x in subscribers_by_channel.index],
)

# Providing axis labels
plt.ylabel("average number of subscribers")
plt.xlabel("channel type")
```

```python
# Clean the dataframe from NaN Latitude and Longitude values
yt.dropna(subset=['Latitude', 'Longitude'], inplace=True)

# Create a base map
m = folium.Map(location=[20,0], zoom_start=2)
marker_cluster = plugins.MarkerCluster().add_to(m)


# Update min and max values
min_subs = yt['subscribers'].min()
max_subs = yt['subscribers'].max()

# Update the colormap
colormap = folium.LinearColormap(colors=['green','yellow','red'], vmin=min_subs, vmax=max_subs)

# Add a color bar
colormap.caption = 'Number of Subscribers'
m.add_child(colormap)

# Plot each YouTube channel using the log-transformed subscriber counts
for index, row in yt.iterrows():
    folium.CircleMarker(
        location=(row['Latitude'], row['Longitude']),
        radius=10,
        color=colormap(row['subscribers']),
        fill=True,
        fill_color=colormap(row['subscribers']),
        tooltip=f"Channel: {row['Youtuber']}<br>Subscribers: {row['subscribers']}"
    ).add_to(marker_cluster)

display(m)
```
